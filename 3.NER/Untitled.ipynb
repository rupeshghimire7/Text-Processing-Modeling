## Named Entity Recognition

```python
nlp = spacy.load("en_core_web_sm")
```

```python
def apply_ner(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities
```

```python
import pandas as pd
```

```python
df['ner_results'] = df['transcript'].apply(apply_ner)
```
